# Data Pre-processing

In this section, we'll explore various data transformation techniques to prepare our dataset for analysis or modeling.

**1. Rescale Data**

Rescaling data is an important step to ensure that all features have a similar range of values. This helps prevent certain features from dominating others in machine learning models. Common rescaling techniques include Min-Max scaling and robust scaling.

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
print("Rescaled Data:")
scaler = MinMaxScaler()
rescaled_data = scaler.fit_transform(dataset)
print(rescaled_data)
```

Rescaling the data scales the features to be in the range [0, 1].

**2. Standardize Data**

Standardizing data involves centering the data around zero and scaling it to have unit variance. This makes the data suitable for algorithms that often assume features are centered around zero and have similar variance. One of the most common standardization methods is the Z-score standardization.

```python
from sklearn.preprocessing import StandardScaler
```

```python
print("Standardized Data:")
# Initialize StandardScaler
scaler = StandardScaler()

# Standardize data
standardized_data = scaler.fit_transform(dataset)
print(standardized_data)
```

**3. Normalize Data**

Normalizing data is similar to rescaling, but it adjusts the range of values to be distributed within a range of [0, 1]. This is often useful in cases where feature values have very different ranges. Normalization can be performed using various methods, such as Min-Max normalization.

```python
from sklearn.preprocessing import Normalizer
```

```python
print("Normalized Data:")
# Initialize Normalizer
scaler = Normalizer()

# Normalize data
normalized_data = scaler.fit_transform(dataset)

print(normalized_data)
```
